```python
data["Company"].value_counts()
```

```python
def add_company(inpt):
    other_brands = {"Samsung", "Razer", "Mediacom", "Microsoft", 
                    "Xiaomi", "Vero", "Chuwi", "Fujitsu", "Google", 
                    "LG", "Huawei"}
    return "Other" if inpt in other_brands else inpt
    
```

```python
data["Company"] = data['Company'].apply(add_company)
```

```python
data["TouchScreen"] = data["ScreenResolution"].apply(lambda x:1 if "Touchscreen" in x else 0)
data["IPS"] = data["ScreenResolution"].apply(lambda x:1 if "IPS" in x else 0)
```

```python
data.head()
```

```python
data["Cpu"].value_counts()
```

```python
data['cpu_name'] = data["Cpu"].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data.head()
```

```python
data["cpu_name"].value_counts()
```

```python
top_cpus = {
    "Intel Core i7", "Intel Core i5", "Intel Core i3",
    "Intel Celeron Dual", "Intel Pentium Quad", "Intel Core M"
}

def categorize_cpu(cpu_name):
    cpu_name = str(cpu_name)
    
    if "AMD" in cpu_name:
        return "AMD"
    
    elif cpu_name in top_cpus:
        return cpu_name
    
    else:
        return "Other"
```

```python
data['cpu_name']= data['cpu_name'].apply(categorize_cpu)
```

```python
data.head(2)
```

```python
data["cpu_name"].value_counts()
```

```python
data["Gpu"].value_counts()
```

```python
data['gpu_name'] = data["Gpu"].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
data.head()
```

```python
data["gpu_name"].value_counts()
```

```python
data = data[data['gpu_name'] != 'ARM']
```

```python
data.shape
```

```python
data["OpSys"].value_counts()
```

```python
def categorize_os(os_name):
    os_name = str(os_name).lower()  
    
    if "windows" in os_name:
        return "Windows"
    elif "mac" in os_name:
        return "Mac"
    elif "linux" in os_name:
        return "Linux"
    elif "chrome" in os_name:
        return "Chrome OS"
    elif "no os" in os_name:
        return "No OS"
    else:
        return "Other"
```

```python
data['OpSys'] = data['OpSys'].apply(categorize_os)
```

```python
data.head()
```

```python
data["OpSys"].value_counts()
```

```python
data = data.drop(columns=['laptop_ID','Inches','Product','ScreenResolution','Cpu','Gpu'])
```

```python
data.head()
```

```python
data = pd.get_dummies(data, dtype=int)
```

```python
data.head()
```

```python
X = data.drop('Price_euros',axis=1)
Y = data['Price_euros']
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.25)
```

```python
X_train
X_train.shape,X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train,y_train)
    acc = model.score(X_test,y_test)
    print(str(model)+ '--->' + str(acc))
```

```python
from sklearn.ensemble import RandomForestRegressor

# Random Forest
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_estimators': [10, 50, 100],
    'criterion': ['squared_error', 'absolute_error', 'poisson']
}
```

```python
grid_obj = GridSearchCV(estimator=rf,param_grid=parameters)
grid_fit = grid_obj.fit(X_train,y_train)
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test,y_test)
```

```python
import pickle
with open('predictor.pickle','wb') as file:
    pickle.dump(best_model,file)
```
